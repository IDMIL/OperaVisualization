```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

```python
y, sr = librosa.load("D:/test_audio/shorttests/short_monkeys.wav")
```

```python
S = np.abs(librosa.stft(y))
chroma = librosa.feature.chroma_stft(S=S, sr=sr)
mfcc = librosa.feature.mfcc(y=y, sr=sr)
fig, ax = plt.subplots(figsize=(8, 12), nrows=5, sharex=True)
img = librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max),
                               y_axis='log', x_axis='time', ax=ax[0])
img = librosa.display.specshow(chroma, y_axis='chroma', x_axis='time', ax=ax[1])
img = librosa.display.specshow(mfcc - max(mfcc.flat), y_axis='linear', x_axis='time', ax=ax[2])
ax[3].plot(np.arange(len(y)) / sr, y)
ax[3].set_ylabel('amplitude')
ax[3].set_xlabel('Time')
ax[4].set_ylabel('energy')
ax[4].set_xlabel('Time')
rms = librosa.feature.rms(y=y)[0]
ax[4].plot(np.arange(len(rms)) * len(y) / (len(rms) * sr), rms)

ax[0].set(title='STFT')
ax[1].set(title='Chroma')
ax[2].set(title='MFCC')
ax[3].set(title='Waveform')
ax[4].set(title='RMS Energy')

for i in range(4):
    ax[i].get_xaxis().set_visible(False)

fig.savefig('representations.svg')
```
